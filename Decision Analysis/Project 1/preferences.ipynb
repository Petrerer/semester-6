```python
import pandas as pd
import subprocess
data = pd.read_csv('Dataset.csv', sep=';')
data = data.reset_index()
data
```

```python
random_pairs = [data.sample(n=2, random_state=i)[['Link']] for i in range(20)]
random_pairs += [data.iloc[[13, 20]][['Link']]]
random_pairs+=[data.iloc[[7, 18]][['Link']]]
random_pairs+=[data.iloc[[8, 15]][['Link']]]
random_pairs+=[data.iloc[[15, 20]][['Link']]]
random_pairs+=[data.iloc[[20, 8]][['Link']]]

global_edge = False
global_dataframe = True

def test_pair(pair_index):
    p = random_pairs[pair_index]

    for url in p["Link"].dropna().unique():
        if global_edge:
            subprocess.Popen(f'start msedge "{url}"', shell=True)

        row = data[data['Link'] == url]
        if global_dataframe:
            display(row)

    return p

preferences = []
```

```python
pair = test_pair(0)
winner = data[data['index'] == 10]['Link'].values[0]
relation = '>' if winner == pair.iloc[0]['Link'] else '<'
preferences.append((pair.iloc[0]['Link'], pair.iloc[1]['Link'], relation))
```
